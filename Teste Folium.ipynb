```python
import sys
sys.path.append('/Users/arthuralvim/Work/folium/')
import json
import numpy as np
import os
import pandas as pd
import string
from sklearn import preprocessing
import folium
```

```python
min_max_scaler = preprocessing.MinMaxScaler()

estados = ['AC', 'AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG',
           'MS', 'MT', 'PA', 'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR',
           'RS', 'SC', 'SE', 'SP', 'TO', ]

estados_br = os.path.join('data', 'brazil-states.geojson.json')
estado_caracteristicas_ = estado_caracteristicas = pd.DataFrame(np.random.randint(0, 1000, size=(1000, 4)), columns=list('ABCD'))

for le in 'ABCD':
    valores = estado_caracteristicas[le].values.reshape(-1, 1)
    valores_normalizados = min_max_scaler.fit_transform(valores)
    estado_caracteristicas[le] = pd.DataFrame(valores_normalizados)
    
print(estado_caracteristicas.shape)
estado_distribuicao = pd.DataFrame(np.random.choice(estados, 1000, replace=True), columns=['UF', ])
print(estado_distribuicao.shape)
estados_dados = pd.concat([estado_distribuicao, estado_caracteristicas], axis=1)
print(estados_dados.shape)
estados_dados.head(5)
```

```python
mapa = folium.Map(width=500, height=550, min_zoom=4, max_zoom=4, location=[-15, -55], 
    prefer_canvas=True, min_lat=-40, max_lat=15, min_lon=-75, max_lon=-30, max_bounds=True,)

mapa.choropleth(
    estados_br,
    data=estados_dados,
    columns=['UF', 'A'],
    key_on='feature.properties.sigla',
    fill_color='RdYlBu',
    fill_opacity=0.9,
    line_opacity=0.2,
    legend_name='Taxa Louca (%)',
    highlight=True
)

#folium.Marker(location=[-15, -55], popup='Centro desse KCT!').add_to(mapa)

folium.BeautifyMarker(location=[-15, -55], popup='Centro desse KCT!', number=30).add_to(mapa)

mapa.save(os.path.join('results', 'brasil-cobertura.html'))
mapa
```
